```python
import os 
import shutil
import random
```

```python
path='/datasets/vision/tiny-imagenet-200'
os.listdir(path)
```

```python
train_path = os.path.join(path,'train')
trn_path=os.path.join(path,'train2')
val_path=os.path.join(path,'val2')
```

```python
def process_cls_id(cls_id,total_cnt):
    # process class cls_id
    tsf=os.path.join(train_path,cls_id)
    tdf=os.path.join(trn_path,cls_id)
    img_names=os.listdir(os.path.join(tsf,'images'))
    cls_box_filename=os.path.join(tsf,cls_id+'_boxes.txt')
    slctd_imgs = random.sample(img_names,50)
    os.makedirs(trn_path,exist_ok=True)
    os.makedirs(os.path.join(tdf,'images'),exist_ok=True)

    os.makedirs(val_path,exist_ok=True)
    vcf=os.path.join(val_path,'images',cls_id)
    os.makedirs(vcf,exist_ok=True)
    val_cnt=0
    # Process the files
    with open(cls_box_filename, "r") as infile, open(os.path.join(val_path,"val_annotations.txt"), "a") as outfile_val, open(os.path.join(tdf,cls_id+"_boxes.txt"), "a") as outfile_trn:
        for line in infile:
            # Split the line into components
            img_name, int1, int2, int3, int4 = line.strip().split("\t")
            # Check if the image name is in our selected strings
            if img_name in slctd_imgs:
                # append the line to outfile_trn
                outfile_trn.write(line)

                dst_filename=f"val_{total_cnt+val_cnt}.JPEG"
                shutil.copy2(os.path.join(tsf,'images',img_name),os.path.join(vcf,dst_filename))
                shutil.copy2(os.path.join(tsf,'images',img_name),os.path.join(tdf,'images',img_name))
                # Create new line with the desired format
                reformatted_line = f"{dst_filename}\t{cls_id}\t{int1}\t{int2}\t{int3}\t{int4}\n"
                outfile_val.write(reformatted_line)
                val_cnt+=1
    return val_cnt
```

```python
trn_cls_folders = os.listdir(train_path)
total_cnt=0
for i,cls_id in enumerate(trn_cls_folders):
    print(f'processing {i+1}/{len(trn_cls_folders)}:{cls_id}')
    curr_cnt=process_cls_id(cls_id,total_cnt)
    total_cnt+=curr_cnt
```
